# Градиентный спуск для функции одной переменной

## Импорт библиотек

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Практика

Найдем локальный минимум функции $f(x)=x^3-5x+2$ методом градиентного спуска.

Сначала определим функцию.

```python
def f(x):
   return x**3 - 5*x + 2
```

Построим график функции.

```python
x_values = [x for x in np.arange(-5, 5, 0.1)]
f_values = [f(x) for x in np.arange(-5, 5, 0.1)]

plt.figure(figsize=(8,8))

plt.axvline(x=0, c = 'black')
plt.axhline(y=0, c = 'black')
plt.plot(x_values, f_values);
```

Вычислим производную функции $f(x)$:

$f'(x)=3x^2-5$.



```python
def grad_f (x):
    return 3*(x**2) - 5
```

Теперь реализуем метод в виде функции.

```python
def gradient_descent(x_start, learning_rate, epsilon, num_iterations):
    x_curr = x_start

    for i in range(num_iterations):
      x_new = x_curr - learning_rate * grad_f(x_curr)

      if abs(x_new - x_curr) < epsilon:
        return x_curr

      x_curr = x_new

    return x_curr
```

Попробуем стартовать из точки $x_0 = 4$ и идти с маленьким шагом $\eta = 0.01$.

```python
x0 = 4
eta = 0.01
epsilon = 0.001
iterations = 10000

x_res = gradient_descent(x0, eta, epsilon, iterations)
x_res
```

Отметим найденную точку на графике.

```python
plt.figure(figsize=(8,8))

plt.axvline(x=0, c = 'black')
plt.axhline(y=0, c = 'black')
plt.plot(x_values, f_values)

plt.scatter(x_res, f(x_res), marker='o', c='red');
```

## Задача

Найдем локальный минимум функции $f(x)=x^4-10x+3$ методом градиентного спуска.

```python
def f(x):
   return x**4 -10*x + 3
```

Построим график функции.

```python
x_values = [x for x in np.arange(-5, 5, 0.1)]
f_values = [f(x) for x in np.arange(-5, 5, 0.1)]

plt.figure(figsize=(8,8))

plt.axvline(x=0, c = 'black')
plt.axhline(y=0, c = 'black')
plt.plot(x_values, f_values);
```

Вычислим производную функции $f(x)$:

$f'(x)=4x^3-10$.


```python
def grad_f (x):
    return 4*(x**3) - 10
```

```python
def gradient_descent(x_start, learning_rate, epsilon, num_iterations):
    x_curr = x_start

    for i in range(num_iterations):
      x_new = x_curr - learning_rate * grad_f(x_curr)

      if abs(x_new - x_curr) < epsilon:
        return x_curr

      x_curr = x_new

    return x_curr
```

```python
x0 = 5
eta = 0.01
epsilon = 0.001
num_iterations = 10000

x_res = gradient_descent(x0, eta, epsilon, num_iterations)
round(x_res, 2)
```

Ради интереса построим график функции.

```python
plt.figure(figsize=(8,8))

plt.axvline(x=0, c = 'black')
plt.axhline(y=0, c = 'black')
plt.plot(x_values, f_values)

plt.scatter(x_res, f(x_res), marker='o', c='red');
```
